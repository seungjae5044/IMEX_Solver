```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torchvision.transforms as T
from torch.utils.data import DataLoader
from PIL import Image
import math
from tqdm.auto import tqdm
import numpy as np
from models import UNet

# Device 설정
device = "cuda" if torch.cuda.is_available() else "cpu"
```

```python
T_steps = 1000
beta_start = 1e-5
beta_end = 0.02
betas = torch.linspace(beta_start, beta_end, T_steps).to(device)  # (T_steps,)
alphas = 1.0 - betas
alphas_cumprod = torch.cumprod(alphas, dim=0)  # 누적곱, (T_steps,)

def q_sample(x0, t, noise=None):
    if noise is None:
        noise = torch.randn_like(x0)
    # 각 t에 대해 sqrt(alpha_bar)와 sqrt(1 - alpha_bar) 계산
    sqrt_alphas_cumprod = torch.sqrt(alphas_cumprod[t]).view(-1, 1, 1, 1)
    sqrt_one_minus_alphas_cumprod = torch.sqrt(1 - alphas_cumprod[t]).view(-1, 1, 1, 1)
    return sqrt_alphas_cumprod * x0 + sqrt_one_minus_alphas_cumprod * noise
```

```python
dataset = []
dataset_size = 5
resolution = 4

for i in range(1,6):
    path_HR = f"data/Set{dataset_size}/image_SRF_{resolution}/img_00{i}_SRF_{resolution}_HR.png"
    path_LR = f"data/Set{dataset_size}/image_SRF_{resolution}/img_00{i}_SRF_{resolution}_LR.png"

    transform = T.Compose([
        T.ToTensor()
        ])
    img_HR = Image.open(path_HR)
    hr_tensor = transform(img_HR).unsqueeze(0).to(device)

    transform_lr = T.Compose([
        T.Resize(tuple(hr_tensor.shape[2:])),
        T.ToTensor()
        ])
    img_LR = Image.open(path_LR).convert("RGB")
    lr_tensor = transform_lr(img_LR).unsqueeze(0).to(device)
    dataset.append(torch.cat([hr_tensor, lr_tensor]))

dataloader = DataLoader(dataset)
```

```python
for X in dataloader:
    print(X.shape)
```

```python
batch_size = 1
model = UNet().to(device=device)
optimizer = torch.optim.Adam(model.parameters(), lr=1e-4)


avg_losses = []
epochs = 1000  # 예시용 epoch, 실제 사용시 더 많은 epoch 필요
model.train()
for epoch in tqdm(range(epochs), desc="Training"):
    losses = []
    for X in dataloader:
        X = X.squeeze()
        optimizer.zero_grad()
        # 배치별 랜덤 time 스텝 선택
        t = torch.randint(0, T_steps, (batch_size,), device=device).long()
        noise = torch.randn_like(X[0].unsqueeze(0))
        # forward diffusion: x_t = sqrt(alpha_bar[t]) * x0 + sqrt(1 - alpha_bar[t]) * noise
        x_t = q_sample(X[0].unsqueeze(0), t, noise)
        x_t = torch.cat([X[0], X[1]], dim=0).unsqueeze(0)
        # 모델 입력: noisy image, time 스텝, 그리고 조건으로 사용되는 LR 이미지
        noise_pred = model(x_t, t)
        loss = F.mse_loss(noise_pred, noise)
        loss.backward()
        optimizer.step()

        losses.append(loss)
    
    avg_loss = np.average(losses)
    avg_losses.append(avg_loss)
    if epoch % 50 == 0:
        print(f"Epoch {epoch}, Loss: {avg_loss.item()}")
```

```python

########################################
# 6. Reverse Diffusion (Sampling) - Super-resolution 생성
########################################
model.eval()
with torch.no_grad():
    # 초기값: pure noise로 시작
    x = torch.randn_like(hr_tensor)
    for t_idx in tqdm(reversed(range(T_steps)), desc="Sampling"):
        t_batch = torch.full((hr_tensor.size(0),), t_idx, device=device, dtype=torch.long)
        x_ = torch.cat([x, lr_tensor], dim=1)
        noise_pred = model(x_, t_batch)
        alpha = alphas[t_idx]
        alpha_bar = alphas_cumprod[t_idx]
        beta = betas[t_idx]
        # reverse step (DDPM 논문 식에 따른 간단한 업데이트)
        if t_idx > 0:
            noise = torch.randn_like(x)
        else:
            noise = 0
        x = (1 / torch.sqrt(alpha)) * (x - ((1 - alpha) / torch.sqrt(1 - alpha_bar)) * noise_pred) + torch.sqrt(beta) * noise

# 결과 이미지 저장 및 표시
output_image = T.ToPILImage()(x.squeeze().clamp(0, 1).cpu())
output_image.save("super_resolution_output.png")
output_image.show()
```
